```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("minisaurus.csv")
```

```python
df.columns
```

```python
df = df[[ 'm', 'z', 'v', 'Myr', 'logL', 'teff', 'density', 'log_R', 'log_g', 'surf_avg_omega_div_omega_crit',
       'surf_avg_v_rot', 'omega', 'Prot', 'profile_number', 'tr_num', 'gyre_fn', 'n1ell0m0', 'n2ell0m0', 'n3ell0m0',
       'n4ell0m0', 'n5ell0m0', 'n6ell0m0', 'n7ell0m0', 'n8ell0m0', 'n9ell0m0',
       'n10ell0m0', 'n1ell1m0', 'n2ell1m0', 'n3ell1m0', 'n4ell1m0', 'n5ell1m0',
       'n6ell1m0', 'n7ell1m0', 'n8ell1m0', 'n9ell1m0', 'n10ell1m0',
       'n1ell1dfreq', 'n2ell1dfreq', 'n3ell1dfreq', 'n4ell1dfreq',
       'n5ell1dfreq', 'n6ell1dfreq', 'n7ell1dfreq', 'n8ell1dfreq',
       'n9ell1dfreq', 'n10ell1dfreq', 'Dnu', 'eps']]
```

```python
df.to_csv('minisaurus.csv')
```

```python
R_ss = np.power(10,df["log_R"])*6.957E08    ##m
omega = df['omega']         ##day-1
M = df['m']*1.989E30        ##kg
G = 6.67408E-11*86400*86400   ##m3 kg-1 day-2
```

```python
R_p = R_ss/(1+(omega**2)*(R_ss**3)/(3*G*M))  ##m
omega_c = np.sqrt(8*G*M/(27*R_p**3))            ##day-1
R_e = R_p*(1+(omega**2)/2)                      ##m
mean_density = 3*M/(4*np.pi*R_e**2*R_p)         ##kg m-3
```

```python
df["omega_c"] = omega_c
df["logR_p"] = np.log10(R_p/6.957E08)
df["logR_e"] = np.log10(R_e/6.957E08)
df["log_mean_density"] = np.log10(mean_density)     
```

```python
df.to_csv("minisaurus_new.csv",index=False)
```
